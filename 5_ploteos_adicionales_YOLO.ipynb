## Generación de ploteos adicionales con los resultados de utilizar conjuntos de entrenamiento con y sin preprocesar sobre la red YOLO prouesta.

### Objetivos:

Generar y analizar las siguientes gráficas a partir de los resultados de la red generados al utilizar el conjunto de imagenes con y sin preprocesamiento:

1. Pérdida de box vs épocas.
2. Pérdida de clases vs épocas.
3. Precisión y Recall vs épocas.
4. mAP50 vs épocas.
5. mAP50-95 vs épocas.

### Características:

En los casos de funciones de pérdida, se muestran tanto los cálculos utilizando el conjunto de entrenamiento como de validación. Mientras que en el caso de las métricas se utilizó solo el conjunto de validación como es de esperar.


```python
# Importa las librerías necesarias
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
# Ruta notebook
file_notebok = os.getcwd()
```

```python
# Nombre del archivo sin preprocesamiento
file_name_rgb = "artefactos/RGB/train_results/results.csv"
# Ruta al archivo sin preprocesamiento
file_path_rgb = os.path.join(file_notebok, file_name_rgb)
```

```python
# Nombre del archivo con preprocesamiento
file_name_prepro = "artefactos/PrePro/train_results/results.csv"
# Ruta al archivo con preprocesamiento
file_path_prepro = os.path.join(file_notebok, file_name_prepro)
```

```python
# Carga el archivo sin preprocesamiento en un DataFrame
df_rgb = pd.read_csv(file_path_rgb)

# Muestra las primeras filas del DataFrame
df_rgb.head()
```

```python
# Carga el archivo con preprocesamiento en un DataFrame
df_prepro = pd.read_csv(file_path_prepro)

# Muestra las primeras filas del DataFrame
df_prepro.head()
```

```python
# PÉRDIDA DE BOXS RGB
# Crea el gráfico
plt.figure(figsize=(10, 6))
# Grafica las curvas
plt.plot(df_rgb["epoch"], df_rgb["train/box_loss"], label="Train Box Loss", marker='o')
plt.plot(df_rgb["epoch"], df_rgb["val/box_loss"], label="Test Box Loss", marker='o')
# Añade título y etiquetas
plt.title("Train vs Test Box Loss - RGB", fontsize=16)
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("Box Loss", fontsize=14)
# Define rangos
plt.xlim(0, 55)
plt.ylim(0, 4.0)
# Añade leyenda
plt.legend(fontsize=12)
# Muestra el gráfico
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# PÉRDIDA DE CLASE RGB
# Crea el gráfico
plt.figure(figsize=(10, 6))
# Grafica las curvas
plt.plot(df_rgb["epoch"], df_rgb["train/cls_loss"], label="Train Cls Loss", marker='o')
plt.plot(df_rgb["epoch"], df_rgb["val/cls_loss"], label="Test Cls Loss", marker='o')
# Añade título y etiquetas
plt.title("Train vs Test Cls Loss - RGB", fontsize=16)
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("Cls Loss", fontsize=14)
# Define rangos
plt.xlim(0, 55)
plt.ylim(0, 4.0)
# Añade leyenda
plt.legend(fontsize=12)
# Muestra el gráfico
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# PÉRDIDA DE BOXS PREPROCESADAS
# Crea el gráfico
plt.figure(figsize=(10, 6))
# Grafica las curvas
plt.plot(df_prepro["epoch"], df_prepro["train/box_loss"], label="Train Box Loss", marker='o')
plt.plot(df_prepro["epoch"], df_prepro["val/box_loss"], label="Test Box Loss", marker='o')
# Añade título y etiquetas
plt.title("Train vs Test Box Loss - PREPROCESADAS", fontsize=16)
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("Box Loss", fontsize=14)
# Define rangos
plt.xlim(0, 55)
plt.ylim(0, 4.0)
# Añade leyenda
plt.legend(fontsize=12)
# Muestra el gráfico
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# PÉRDIDA DE CLASE PREPROCESADAS
# Crea el gráfico
plt.figure(figsize=(10, 6))
# Grafica las curvas
plt.plot(df_prepro["epoch"], df_prepro["train/cls_loss"], label="Train Cls Loss", marker='o')
plt.plot(df_prepro["epoch"], df_prepro["val/cls_loss"], label="Test Cls Loss", marker='o')
# Añade título y etiquetas
plt.title("Train vs Test Class Loss - PREPROCESADAS", fontsize=16)
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("Cls Loss", fontsize=14)
# Define rangos
plt.xlim(0, 55)
plt.ylim(0, 4.0)
# Añade leyenda
plt.legend(fontsize=12)
# Muestra el gráfico
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# PRECISION
# Crea el gráfico
plt.figure(figsize=(10, 6))
# Grafica las curvas
plt.plot(df_rgb["epoch"], df_rgb["metrics/precision(B)"], label="Sin procesar Precisión", marker='o')
plt.plot(df_prepro["epoch"], df_prepro["metrics/precision(B)"], label="Prepocesadas Precisión", marker='o')
# Añade título y etiquetas
plt.title("Precisión Imagenes sin procesar vs Procesadas", fontsize=16)
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("Precisión", fontsize=14)
# Define rangos
plt.xlim(0, 55)
plt.ylim(0, 1.2)
# Añade leyenda
plt.legend(fontsize=12)
# Muestra el gráfico
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# RECALL
plt.figure(figsize=(10, 6))
# Grafica las curvas
plt.plot(df_rgb["epoch"], df_rgb["metrics/recall(B)"], label="Sin procesar Recall", marker='o')
plt.plot(df_prepro["epoch"], df_prepro["metrics/recall(B)"], label="Prepocesadas Recall", marker='o')
# Añade título y etiquetas
plt.title("Recall Imagenes sin procesar vs Procesadas", fontsize=16)
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("Recall", fontsize=14)
# Define rangos
plt.xlim(0, 55)
plt.ylim(0, 1.2)
# Añade leyenda
plt.legend(fontsize=12)
# Muestra el gráfico
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# METRICA MAP50
# Crea el gráfico
plt.figure(figsize=(10, 6))
# Grafica las curvas
plt.plot(df_rgb["epoch"], df_rgb["metrics/mAP50(B)"], label="Sin procesar mAP50", marker='o')
plt.plot(df_prepro["epoch"], df_prepro["metrics/mAP50(B)"], label="Prepocesadas mAP50", marker='o')
# Añade título y etiquetas
plt.title("mAP50 Imagenes sin procesar vs Procesadas", fontsize=16)
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("mAP50", fontsize=14)
# Define rangos
plt.xlim(0, 55)
plt.ylim(0, 1.2)
# Añade leyenda
plt.legend(fontsize=12)
# Muestra el gráfico
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# METRICA MAP50-95
plt.figure(figsize=(10, 6))
# Grafica las curvas
plt.plot(df_rgb["epoch"], df_rgb["metrics/mAP50-95(B)"], label="Sin procesar mAP50-95", marker='o')
plt.plot(df_prepro["epoch"], df_prepro["metrics/mAP50-95(B)"], label="Prepocesadas mAP50-95", marker='o')
# Añade título y etiquetas
plt.title("mAP50-95 Imagenes sin procesar vs Procesadas", fontsize=16)
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("mAP50-95", fontsize=14)
# Define rangos
plt.xlim(0, 55)
plt.ylim(0, 1.2)
# Añade leyenda
plt.legend(fontsize=12)
# Muestra el gráfico
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

### Observaciones:

A partir de la comparación de utilizar el conjunto de datos con y sin preprocesamiento surgieron los siguientes resultados:

El conjunto de datos sin preprocesamiento obtuvo entre un 0.80 y 0.90 de mAP50-95 en todas las épocas (una vez sorteadas las primeras épocas de estabilización), mientras que con el preprocesamiento propuesto la mAP50-95 estuvo entre los 0.60 y 0.70 en ell mimos rango de épocas.

Además, en cuanto al Recall y Presicion, nuevamente se obtuvo valores del orden de 0,95 (salvo en las primeras épocas) en el caso sin preprocesamiento, mientras que con el preprocesamiento propuesto estas métricas rondaron los 0,70 y 0,90 respectivamente en el mismo rango de épocas.

En cuanto a las pérdidas de entrenamiento y prueba para la detección (box) y clasificación, no se encontró sobreajuste o subajuste en ningún caso. No obstante, en el caso de conjunto de entrenamiento con imagenes sin preprocesar, la generalización resultó mejor dado que las curvas de train y test convergen al mínimo. Mientras que en el caso con preprocesamiento hay diferencias, especialmente en el caso de clasificación donde la pérdida de prueba se mantiene significativamente mayor a la pérdida de entrenameinto.
